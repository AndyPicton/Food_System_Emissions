```python
# Import dependencies for
import pandas as pd
```

```python
# Import data
sector_emi_df = pd.read_csv("FOOD emi by sector.csv")  
sector_emi_df
```

```python
sector_emi_df.dtypes
```

```python
# Find null values
for column in sector_emi_df.columns:
    print(f"Column {column} has {sector_emi_df[column].isnull().sum()} null values")
```

```python
# Drop the null rows
sector_emi_df = sector_emi_df.dropna()
```

```python
sector_emi_df
```

```python
# Confirm all NA values are gone
sector_emi_df.isnull().sum(axis = 1)
```

```python
# Change column names
sector_emi_df.rename(columns={"Country_code_A3": "Country Code", "Name": "Country", "C_group_IM24_sh": "Region", "dev_country": "Dev or Ind", "FOOD_system_stage": "Food System Stage"})
```
